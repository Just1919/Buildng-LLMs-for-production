## рҹӨ– LangChainвҖҷs Role in Retrieval-Augmented Generation (RAG)

**Retrieval-Augmented Generation (RAG)** is a powerful technique designed to address one of the main limitations of Large Language Models (LLMs): **hallucinations**.

By integrating external knowledge sources, RAG allows LLMs to access **relevant and factual information** during the generation process.  
This leads to outputs that are:
- More accurate  
- More reliable  
- Better adapted to the given context  

---

## рҹ§  Why RAG is Important

LLMs generate responses based on patterns learned during training, which can sometimes lead to incorrect or fabricated information.  

рҹ‘ү RAG solves this by:
- Retrieving real data from external sources (documents, databases, etc.)
- Injecting this information into the model before generating a response

---

## вҡҷпёҸ LangChain for RAG

LangChain provides powerful abstractions to easily build RAG systems.

With its retrieval components, developers can:
- Connect LLMs to external data sources

рҹ‘ү This enables LLM applications to produce:
- More accurate answers

## рҹ§© Key LangChain Concepts & Components

- **Prompts**  
  LangChain provides tools to create and manage prompt templates.  
  These templates act as predefined structures for generating inputs to language models.

- **Output Parsers**  
  Output parsers help structure the responses generated by LLMs.  
  They transform raw outputs into more organized and usable formats.

- **Retrievers**  
  Retrievers take a query as input and return a list of relevant documents.  
  LangChain supports advanced retrieval methods and integrates with external retrieval systems.

- **Document Loaders**  
  A document consists of text and its associated metadata.  
  Document loaders allow you to load data from various sources into this structured format.

- **Text Splitters**  
  These tools divide large texts into smaller chunks.  
  LangChain provides built-in utilities to split, merge, and filter documents efficiently.

- **Indexes**  
  Indexes are data structures used to organize and store information for fast and efficient search.

- **Embedding Models**  
  Embedding models convert text into numerical vectors.  
  LangChain provides a unified interface for multiple providers such as OpenAI, Cohere, and Hugging Face.

- **Vector Stores**  
  Vector stores store embeddings and enable similarity search.  
  They are essential for handling and querying unstructured data.

- **Agents**  
  Agents are decision-making components that determine the sequence of actions to solve a task.

- **Chains**  
  Chains represent sequences of operations (LLMs, tools, or data processing steps).  
  They combine multiple components into a cohesive workflow.

- **Tools**  
  Tools are functions that allow the model to access external resources, such as:
  - Web search  
  - Databases  
  - Python execution environments  

- **Memory**  
  Memory stores previous interactions to provide context for future responses.

- **Callbacks**  
  Callbacks allow you to monitor and interact with different stages of your application.  
  They are useful for logging, debugging, and streaming outputs.

## рҹӨ– LangChain Agents & Tools Overview

### рҹ”Қ What are Agents?

LangChain agents are components designed to **automatically complete tasks** by combining:
- Chains  
- Prompts  
- Memory  
- Tools  

These agents can:
- Execute multi-step workflows  
- Interact with external systems (e.g., Gmail, SQL databases)  
- Dynamically decide how to solve a task  

рҹ‘ү They act as **decision-makers**, determining which actions to take and which tools to use.

---

## вҡҷпёҸ Agent Types in LangChain

LangChain provides several types of agents, each tailored for specific use cases:

- **Zero-shot ReAct Agent**  
  Uses the ReAct framework to decide which tools to use based only on their descriptions.  
  No prior examples are needed.

- **Structured Input ReAct Agent**  
  Designed for tools that require multiple inputs.  
  Suitable for more complex interactions.

- **OpenAI Functions Agent**  
  Built for function calling with advanced models such as:
  - gpt-3.5-turbo  
  - gpt-4-turbo  

- **Self-Ask with Search Agent**  
  Focuses on answering factual questions by:
  - Breaking them into sub-questions  
  - Using search tools to find intermediate answers  

- **ReAct Document Store Agent**  
  Combines:
  - Search  
  - Lookup  
  To maintain a continuous reasoning process over documents.

- **Plan-and-Execute Agent**  
  Creates a plan with multiple steps and executes them sequentially.  
  Particularly useful for:
  - Complex tasks  
  - Long workflows  

  вҡ пёҸ Trade-off: May introduce higher latency.

### Summarization Chain Example


A summarization chain is a process in LangChain that enables automatic generation of concise summaries from large amounts of text by structuring the interaction with a language model.

In this example, the summarization chain is used to process an external document (a PDF) and produce a clear and structured summary. The goal is to load the document, extract its content, and pass it to a language model for summarization.

To do this, the language model is initialized with ChatOpenAI and a temperature of 0, ensuring deterministic and consistent outputs. Then, the PyPDFLoader class is used to load the PDF file page by page into a format that can be easily processed. All pages are then combined into a single text string.

A prompt is defined using PromptTemplate to instruct the model to summarize the content clearly and concisely. The chain is created using LangChain Expression Language (prompt | llm) and executed with the invoke() method.

For this example, the PDF used is the Python  course , downloaded from the https://static.realpython.com/python-cheatsheet.pdf.

It is important to install the pypdf package before running this code.


````python
#Import necesseray modules
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_community.document_loaders import PyPDFLoader
````

````python
!pip install pypdf
````

````python
# Initialize language model
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)

# Load the document using PyPDFLoader
document_loader = PyPDFLoader("/content/drive/MyDrive/llms_for_production/python-cheatsheet.pdf")
documents = document_loader.load()

# Combine all pages into one text
text = "\n\n".join(doc.page_content for doc in documents)

# Create the prompt template
prompt = PromptTemplate.from_template(
    "Summarize the following PDF content in a clear and concise way:\n\n{text}"
)

# Create the chain
chain = prompt | llm

# Summarize the document
response = chain.invoke({"text": text})

# Print the summary
print(response.content)
````

When using this model, pay attention to the size of the PDF. The maximum context length is 16,385 tokens. If this limit is exceeded, it may result in an error (context_length_exceeded). In such cases, it is recommended to use another approach, such as splitting the document into smaller chunks and applying a summarization strategy like map-reduce or refine.

## QA Chain Example

LangChain allows prompts to be structured in different ways, including asking general questions to language models.

вҡ пёҸ However, it is important to be aware of potential hallucinations, where the model may generate incorrect or non-factual information. To address this issue, a Retrieval-Augmented Generation (RAG) approach can be implemented. In the next article, we will explore how LangChain facilitates the implementation of such systems using a Retrieval Chain.

In this example, we define a custom prompt template using the PromptTemplate class. The template includes a {question} placeholder, which represents the userвҖҷs input, followed by a newline and the label вҖңAnswer:вҖқ.

The input_variables parameter specifies the variables expected in the templateвҖ”in this case, the question. These variables are dynamically replaced with actual input values when the chain is executed.

````python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI

# Create the prompt with input_variables
prompt = PromptTemplate(
    template="Question: {question}\nAnswer:",
    input_variables=["question"]
)
````

Next, an instance of the OpenAI model gpt-3.5-turbo is initialized with a temperature set to 0, ensuring fully deterministic outputs. This model is created using the ChatOpenAI class, where the model and temperature parameters are specified.

After that, a question-answering chain is constructed by combining the language model with the previously defined prompt template. In earlier versions of LangChain, this was typically done using the LLMChain class, which takes two main arguments: the language model (llm) and the prompt template (prompt).

In modern LangChain usage, this process is simplified by directly composing the prompt and the model using the expression:

````python
# Initialize the model
llm = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0
)

# Create the chain
chain = prompt | llm
````

By following these steps, input questions can be processed efficiently using the custom question-answering chain. This approach enables the generation of relevant and coherent answers by combining the capabilities of the OpenAI model with a well-defined prompt template.

````python

# Invoke the chain
response = chain.invoke({"question": "does money bring happiness?"})

print(response.content)
````

This example illustrates how LangChain facilitates the integration of prompt templates in question-answering applications. The framework can be further extended by incorporating additional components such as retrieval-augmented generation, agents, or memory mechanisms, enabling the development of more advanced and intelligent applications.

We selected a set of representative news article URLs in order to generate summaries. The following code demonstrates how to use the requests library to fetch the content of each URL while specifying a custom User-Agent header to mimic a real browser.

The newspaper3k library is then used to parse each article and extract its title and main text content.

````python
import requests
from newspaper import Article

# Define headers to mimic a real browser (helps avoid blocking)
headers = {
    'User-Agent': (
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
        'AppleWebKit/537.36 (KHTML, like Gecko) '
        'Chrome/120.0.0.0 Safari/537.36'
    )
}

# Target article URL
article_url ="https://blog.python.org/2026/03/the-python-insider-blog-has-moved/"



# Create a session object for better performance and connection reuse
session = requests.Session()

try:
    # Send a GET request with headers and timeout
    response = session.get(article_url, headers=headers, timeout=10)

    # Check if the request was successful (HTTP 200 OK)
    if response.status_code == 200:

        # Extract the raw HTML content from the response
        html = response.text

        # Initialize the Article object with the URL
        article = Article(article_url)

        # IMPORTANT: Set the HTML manually to avoid a second request
        article.set_html(html)

        # Parse the article content (title, text, etc.)
        article.parse()

        # Print extracted information
        print("Title:", article.title)
        print("Text:", article.text)

    else:
        # Print error code if request failed
        print(f"Failed to fetch article (status code: {response.status_code})")

# Handle timeout errors
except requests.exceptions.Timeout:
    print("Request timed out")

# Handle connection errors (DNS issues, no internet, etc.)
except requests.exceptions.ConnectionError:
    print("Connection error occurred")

# Catch any other unexpected errors
except Exception as e:
    print(f"Unexpected error: {e}")

# Always close the session to free resources
finally:
    session.close()
````
