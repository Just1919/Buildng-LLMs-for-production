## Building LLM-Powered Applications with LangChain


LangChain is an open-source framework that makes it easier to build, scale, and deploy applications powered by large language models (LLMs). It offers modular components, tools, and integrations that streamline every step of the LLM application lifecycle.

## 📌 Key Features of LangChain

- **Abstractions & LCEL (LangChain Expression Language)**  
  Enable flexible and modular composition of chains using a clean and modern syntax.

- **Third-party integrations & partner packages**  
  Provide easy extensibility by connecting with external tools and services.

- **Chains, Agents & Retrieval strategies**  
  Allow building intelligent systems capable of reasoning, decision-making, and information retrieval.

- **LangGraph**  
  Designed for creating robust and stateful multi-actor applications.

- **LangServe**  
  Used to deploy LangChain applications as REST APIs for production environments.

---

## 🌐 LangChain Ecosystem

- **LangSmith**  
  A developer platform for:
  - Debugging
  - Testing
  - Evaluating
  - Monitoring LLM applications

## 🤖 LangChain’s Role in Retrieval-Augmented Generation (RAG)

**Retrieval-Augmented Generation (RAG)** is a powerful technique designed to address one of the main limitations of Large Language Models (LLMs): **hallucinations**.

By integrating external knowledge sources, RAG allows LLMs to access **relevant and factual information** during the generation process.  
This leads to outputs that are:
- More accurate  
- More reliable  
- Better adapted to the given context  

---

## 🧠 Why RAG is Important

LLMs generate responses based on patterns learned during training, which can sometimes lead to incorrect or fabricated information.  

👉 RAG solves this by:
- Retrieving real data from external sources (documents, databases, etc.)
- Injecting this information into the model before generating a response

---

## ⚙️ LangChain for RAG

LangChain provides powerful abstractions to easily build RAG systems.

With its retrieval components, developers can:
- Connect LLMs to external data sources  
- Query documents or databases efficiently  
- Inject relevant context into prompts  

👉 This enables LLM applications to produce:
- More accurate answers  
- Context-aware responses  
- Knowledge-grounded outputs

## 🧩 Key LangChain Concepts & Components

- **Prompts**  
  LangChain provides tools to create and manage prompt templates.  
  These templates act as predefined structures for generating inputs to language models.

- **Output Parsers**  
  Output parsers help structure the responses generated by LLMs.  
  They transform raw outputs into more organized and usable formats.

- **Retrievers**  
  Retrievers take a query as input and return a list of relevant documents.  
  LangChain supports advanced retrieval methods and integrates with external retrieval systems.

- **Document Loaders**  
  A document consists of text and its associated metadata.  
  Document loaders allow you to load data from various sources into this structured format.

- **Text Splitters**  
  These tools divide large texts into smaller chunks.  
  LangChain provides built-in utilities to split, merge, and filter documents efficiently.

- **Indexes**  
  Indexes are data structures used to organize and store information for fast and efficient search.

- **Embedding Models**  
  Embedding models convert text into numerical vectors.  
  LangChain provides a unified interface for multiple providers such as OpenAI, Cohere, and Hugging Face.

- **Vector Stores**  
  Vector stores store embeddings and enable similarity search.  
  They are essential for handling and querying unstructured data.

- **Agents**  
  Agents are decision-making components that determine the sequence of actions to solve a task.

- **Chains**  
  Chains represent sequences of operations (LLMs, tools, or data processing steps).  
  They combine multiple components into a cohesive workflow.

- **Tools**  
  Tools are functions that allow the model to access external resources, such as:
  - Web search  
  - Databases  
  - Python execution environments  

- **Memory**  
  Memory stores previous interactions to provide context for future responses.

- **Callbacks**  
  Callbacks allow you to monitor and interact with different stages of your application.  
  They are useful for logging, debugging, and streaming outputs.

## 🤖 LangChain Agents & Tools Overview

### 🔍 What are Agents?

LangChain agents are components designed to **automatically complete tasks** by combining:
- Chains  
- Prompts  
- Memory  
- Tools  

These agents can:
- Execute multi-step workflows  
- Interact with external systems (e.g., Gmail, SQL databases)  
- Dynamically decide how to solve a task  

👉 They act as **decision-makers**, determining which actions to take and which tools to use.

---

## ⚙️ Agent Types in LangChain

LangChain provides several types of agents, each tailored for specific use cases:

- **Zero-shot ReAct Agent**  
  Uses the ReAct framework to decide which tools to use based only on their descriptions.  
  No prior examples are needed.

- **Structured Input ReAct Agent**  
  Designed for tools that require multiple inputs.  
  Suitable for more complex interactions.

- **OpenAI Functions Agent**  
  Built for function calling with advanced models such as:
  - gpt-3.5-turbo  
  - gpt-4-turbo  

- **Self-Ask with Search Agent**  
  Focuses on answering factual questions by:
  - Breaking them into sub-questions  
  - Using search tools to find intermediate answers  

- **ReAct Document Store Agent**  
  Combines:
  - Search  
  - Lookup  
  To maintain a continuous reasoning process over documents.

- **Plan-and-Execute Agent**  
  Creates a plan with multiple steps and executes them sequentially.  
  Particularly useful for:
  - Complex tasks  
  - Long workflows  

  ⚠️ Trade-off: May introduce higher latency.

---

## 🧠 Key Idea

Agents are responsible for:
- Selecting the appropriate actions  
- Deciding whether to use:
  - One tool  
  - Multiple tools  
  - Or no tool at all  

👉 They provide the **reasoning and orchestration layer** in LangChain applications.

## 💬 Prompt Templates in LangChain

LangChain uses **message prompt templates** to construct and manage prompts effectively, enabling better utilization of chat-based models.

Different types of prompts serve specific roles:

- **SystemMessagePromptTemplate**  
  Provides initial instructions, context, or rules to guide the AI model’s behavior.

- **HumanMessagePromptTemplate**  
  Represents user inputs or questions that the AI model needs to answer.

````python
#Import the necessary modules
from langchain_openai import ChatOpenAI
from langchain_core.prompts import (
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate,
)
````

````python
# Before executing the following code, make sure to have
# your OpenAI key saved in the "OPENAI_API_KEY" environment variable.
chat = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0)
````

````python
template="You are an assistant that helps users find information about movies."
system_message_prompt = SystemMessagePromptTemplate.from_template(template)
human_template="Find information about the movie or series TV {movie_title}."
human_message_prompt = HumanMessagePromptTemplate.from_template(human_template)

chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])

reponse = chat.invoke(
    chat_prompt.format_prompt(movie_title="Silicon Valley").to_messages()
)

print(reponse.content)
````

### Summarization Chain Example


````python
#Import necesseray modules
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_community.document_loaders import PyPDFLoader
````

````python
!pip install pypdf
````

````python
# Initialize language model
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)

# Load the document using PyPDFLoader
document_loader = PyPDFLoader("/content/drive/MyDrive/Formation_Talend_DI/00 Course Introduction_FR.pdf")
documents = document_loader.load()

# Combine all pages into one text
text = "\n\n".join(doc.page_content for doc in documents)

# Create the prompt template
prompt = PromptTemplate.from_template(
    "Summarize the following PDF content in a clear and concise way:\n\n{text}"
)

# Create the chain
chain = prompt | llm

# Summarize the document
response = chain.invoke({"text": text})

# Print the summary
print(response.content)
````
