```python
# use TextLoader to load text from local file
from langchain_community.document_loaders import TextLoader
loader=TextLoader('my_file.txt')
docs_from_file=loader.load()

print("Number of documents :", len(docs_from_file))
print("First documents: ",docs_from_file)
```

```python
print("First documents: ",docs_from_file[0])
```

2. Use CharacterTextSplitter to split the documents into text
snippets called “chunks.” Chunk_overlap is the number of
characters that overlap between two chunks. It preserves context
and improves coherence by ensuring that important information
is not cut off at the boundaries of chunks.

```python
from langchain_text_splitters import CharacterTextSplitter, RecursiveCharacterTextSplitter
#Create a text splitter
text_spliter=RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=20)

#Split documents into chunk
docs=text_spliter.split_documents(docs_from_file)
```

```python
print(len(docs))
```

```python
print(docs)
```

3. Create a vector embedding for each text snippet.

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-ada-002")
```

```python
#!pip install --upgrade --quiet langchain-openai langchain-deeplake tiktoken
```

```python
from langchain_deeplake.vectorstores import DeeplakeVectorStore
```

```python
#Create a vector store: Deep Lake

my_activeloop_dataset_name="langchain_course_indexers_retrievers"
my_activeloop_org_id="justindjakas4"

dataset_path=f"hub://{my_activeloop_org_id}/{my_activeloop_dataset_name}"


db = DeeplakeVectorStore(dataset_path=dataset_path,embedding_function=embeddings)

#Add documents to the Deep Lake dataset
db.add_documents(docs)
```

4. The next is creating a LangChain retriever by calling the
.as_retriever() method on your vector store instance.

```python
#Create retriever from db
retriever=db.as_retriever()
```

5. Once we have the retriever, we can use the RetrievalQA class to
define a question answering chain using external data source and
start with question-answering.

```python
from langchain.chains import RetrievalQA
from langchain_openai import OpenAI,ChatOpenAI
```

```python
#Create a retriaval chain
llm = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0
)

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever
)
```

5. We can query our document about a specific topic found in the documents.

```python
query=" How google plans to challenge OpenAI"
response=qa_chain.invoke(query)
print(response["result"])
```

6. ChatBot

```python
import time
import sys

# Function to simulate a "thinking..." animation
def thinking_animation():
    print("\nChatbot is thinking", end="")  # Start on a new line
    for _ in range(3):
        time.sleep(0.5)  # Wait for half a second
        print(".", end="")  # Print dots progressively
        sys.stdout.flush()  # Force immediate display
    print("\n")  # Add spacing after animation


def chatbot():
    # Display a welcome message when the chatbot starts
    print("Chatbot ready. Type 'quit' to exit.\n")

    # Infinite loop to keep the chatbot running
    while True:
        # Take user input and remove extra spaces
        query = input("You: ").strip()

        # Check if the user wants to exit the chatbot
        if query.lower() in ["quit", "exit", "bye"]:
            print("\nChatbot: Goodbye!\n")
            break  # Exit the loop → stop the chatbot

        # Check if the user entered an empty message
        if not query:
            print("\nChatbot: Please enter a question.\n")
            continue  # Skip the rest and ask for input again

        try:
            # Add space before chatbot response
            print()

            # Show thinking animation
            thinking_animation()

            # Send the user's query to the RetrievalQA chain
            response = qa_chain.invoke(query)

            # Display the generated answer with spacing
            print(f"Chatbot: {response['result']}\n")

        except Exception as e:
            # Handle errors (API issues, model problems, etc.)
            print(f"\nChatbot: An error occurred: {e}\n")


# Start the chatbot
chatbot()
```

##ContextualCompressionRetriever with the LLMChainExtractor:

```python
from langchain.retrievers import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import LLMChainExtractor

#Create GPT3
llm = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0
)

#Create compressor for the retriever

compressor=LLMChainExtractor.from_llm(llm)
compression_retriever=ContextualCompressionRetriever(base_compressor=compressor,base_retriever=retriever)
```

Once the compression_retriever is created, we can retrieve the relevant compressed documents for a query.

```python
# retrieving compressed documents
retrieved_docs = compression_retriever.invoke(
 "How Google plans to challenge OpenAI?"
)
print(retrieved_docs[0].page_content)
```

```python
query = "How Google plans to challenge OpenAI?"

retrieved_docs = compression_retriever.invoke(query)

context = "\n\n".join([doc.page_content for doc in retrieved_docs])

prompt = f"""
Answer the following question based only on the provided context.

Context:
{context}

Question:
{query}
"""

response = llm.invoke(prompt)

print(response.content)
```
