```python
# Install the Wikipedia reader if not already installed
# !pip install llama-index-readers-wikipedia

from llama_index.readers.wikipedia import WikipediaReader

# Create an instance of the Wikipedia reader
loader = WikipediaReader()

# Load data from specific Wikipedia pages
documents = loader.load_data(
    pages=["Large Language Models", "Natural Language Processing"]
)

# Print the number of documents retrieved
print(f"Number of documents: {len(documents)}")

# Preview the first document
print(documents[0].text[:500])
```

##Nodes

```python
from llama_index.core.node_parser import  SimpleNodeParser
#Assuming documents have already been loaded
#Initialise the parser
parser=SimpleNodeParser.from_defaults(chunk_size=512, chunk_overlap=20)
#Parse the documents into nodes
nodes=parser.get_nodes_from_documents(documents)
print(len(nodes))
```
