# Tanslation with LLMs(GPT-3.5 API)

```python
from dotenv import load_dotenv
import os
import openai

# Replace the path with the full path to your .env file
load_dotenv('/content/drive/MyDrive/llms_for_production/.env')

# Access the keys as environment variables
openai_key = os.getenv("OPENAI_API_KEY")
deeplake_token = os.getenv("DEEPLAKE_TOKEN")

print("Keys loaded:", bool(openai_key), bool(deeplake_token))
```

```python
#Import  librairies
from openai import OpenAI
client=OpenAI(api_key=os.environ['OPENAI_API_KEY'])


english_text = "Hello how are you doing?"

response = client.chat.completions.create(
    model="gpt-3.5-turbo",  # ou gpt-3.5-turbo si dispo
messages=[
        {"role":"system","content":"You are a helpful assistant that translates English to French."},
        {"role":"user","content":f"Translate the following English text to French: '{english_text}'"}
    ],
)

print(response.choices[0].message.content)

```

# Control LLMs Output by providing Examples: Few-shot learning

```python
#Prompt for summarization
prompt="""
Describing the following movie using emojis.
{movie}: """

examples = [
    { "input": "Titanic", "output": " â¤ " },
    { "input": "The Matrix", "output": "  " }
]

movie="Toy Story"

response=client.chat.completions.create(
    model="gpt-3.5-turbo",
     messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt.format(movie=examples[0]["input"])},
        {"role": "assistant", "content": examples[0]["output"]},
        {"role": "user", "content": prompt.format(movie=examples[1]["input"])},
        {"role": "assistant", "content": examples[1]["output"]},
        {"role": "user", "content": prompt.format(movie=movie)},
  ]
)

print(response.choices[0].message.content)
```

# Text-to-SQL Engine

```python

# Initialize OpenAI client
client = OpenAI(api_key=os.environ['OPENAI_API_KEY'])

# Example text to convert to SQL
text_to_convert = "Get all customers who made a purchase in the last 30 days."

# Send request to the model
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant that converts English text into SQL queries."
        },
        {
            "role": "user",
            "content": f"Convert the following English text into an SQL query: '{text_to_convert}'"
        }
    ],
)

# Print the generated SQL query
print(response.choices[0].message.content)
```

```python
from openai import OpenAI
import os

# Load your API key from environment variable
client = OpenAI(api_key=os.environ['OPENAI_API_KEY'])

def text_to_sql_chatbot():
    print("Welcome to the Text-to-SQL Chatbot! Type 'exit' to quit.\n")
    while True:
        user_input = input("User: ")
        if user_input.lower() == "exit":
            print("Goodbye!")
            break

        # Call OpenAI API
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {
                    "role": "system",
                    "content": "You are a helpful assistant that converts natural language questions into SQL queries."
                },
                {
                    "role": "user",
                    "content": user_input
                }
            ]
        )

        sql_query = response.choices[0].message.content
        print(f"SQL Query:\n{sql_query}\n")

# Run chatbot
text_to_sql_chatbot()
```
